```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
r = np.linspace(0, 0.03, 30)
theta = np.linspace(0, 2*np.pi, 1000)
```

```python
for i in r:
    plt.scatter(i*np.cos(theta),i*np.sin(theta), c = 'blue', s = 2)
```

```python
T = np.zeros((len(r), len(theta)))
```

```python
dr = r[1]-r[0]
dth = theta[1]-theta[0]
alph = 1e-7
```

```python
dt = 1

N = len(theta)
T[:,:] = 10
T[len(r)-1,:] = 180
Tp = np.copy(T)
rad = np.zeros(T.shape)
for i in range(len(r)):
    rad[i,:] = r[i]


for t in range(100000):
    Tp[1:-1, 1:-1] = T[1:-1,1:-1] + (alph*dt/(rad[1:-1,1:-1]*dr))*((T[2:,1:-1]-T[0:-2,1:-1])/dr) + \
    (alph*dt/dr**2)*(T[2:,1:-1]-2*T[1:-1,1:-1]+T[:-2,1:-1]) + \
    alph*dt/(dth**2*rad[1:-1,1:-1]**2)*(T[1:-1,:-2]-2*T[1:-1,1:-1]+T[1:-1,2:])
    
    #Accounting for theta = 0
    Tp[1:-1,0] = T[1:-1,0] + (alph*dt/(rad[1:-1,0]*dr))*((T[2:,0]-T[0:-2,0])/dr) + \
    (alph*dt/dr**2)*(T[2:,0]-2*T[1:-1,0]+T[:-2,0]) + \
    alph*dt/(dth**2*rad[1:-1,0]**2)*(T[1:-1,1]-2*T[1:-1,0]+T[1:-1,-2])
    
    #Accounting for theta = 2pi
    Tp[1:-1,-1] =T[1:-1,0] 
    
    #Accounting for r = 0
    Tp[0, :] = np.average(T[1,:])
    T = np.copy(Tp)
```

```python
theta_matrix, radius_matrix = np.meshgrid(theta,r)
```

```python
X = radius_matrix * np.cos(theta_matrix)
Y = radius_matrix * np.sin(theta_matrix)
```

```python
plt.contourf(X,Y,T)
plt.colorbar()
```
